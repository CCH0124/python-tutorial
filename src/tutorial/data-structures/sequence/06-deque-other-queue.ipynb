在 list 頭插入和移除元素的代價很高，因為記憶體中整個 list 都要移動。

`collections.deque` 類別是*執行續安全* 的雙頭佇列。

depue 外，其它還有
- queue
  - 執行緒安全
- multiprocessing
  - 實作 `SimpleQueue` 與 `Queue`
  - 為了進行程序之間的溝通而設計
- asyncio
  - 非同步程式中管理任務
- heapq
  - 使可變的 sequence 作為 heap 佇列或優先順序佇列


```python
from collections import deque
```

```python
dq = deque(range(10), maxlen=10) # maxlen 佇列最多可容納項目數
dq
```

```python
dq.rotate(3) # > 0 會從右端取出，並將它們加入左端。 < 0 時反之。
dq
```

```python
dq.rotate(-4)
dq
```

```python
dq.appendleft(-1)
dq
```

```python
dq.extend([11, 22, 33]) # 從右端加入元素，並將左端元素彈出，保持 maxlen 數量
dq
```

```python
dq.extendleft([10, 20, 30, 40])
dq
```
